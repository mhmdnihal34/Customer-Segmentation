```python
import pandas as pd
data=pd.read_csv('output.csv')
data
```

```python
data.isnull().sum()
```

```python
data=data[~data['InvoiceNo'].astype(str).str.startswith('c')]
```

```python
data=data[(data['Quantity']>0) & (data['UnitPrice']>0)]
```

```python
data['TotalAmount']=data['Quantity'] * data['UnitPrice']
data
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data
```

```python
data.describe()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Quantity'], bins=50)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=data['UnitPrice'])
plt.show()
```

```python
top_customers = data.groupby('CustomerID')['TotalAmount'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
country_sales = data.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).head(10)
country_sales.plot(kind='bar', figsize=(8,4))
plt.show()
```

```python
sns.heatmap(data[['Quantity','UnitPrice','TotalAmount']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
data['Month'] = data['InvoiceDate'].dt.to_period('M')
monthly_sales = data.groupby('Month')['TotalAmount'].sum()
monthly_sales.plot(figsize=(8,4))
plt.show()
```
