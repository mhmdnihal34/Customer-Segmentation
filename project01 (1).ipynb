```python
import pandas as pd
data=pd.read_csv('output.csv')
data
```

```python
data.isnull().sum()
```

```python
data=data[~data['InvoiceNo'].astype(str).str.startswith('c')]
```

```python
data=data[(data['Quantity']>0) & (data['UnitPrice']>0)]
```

```python
data['TotalAmount']=data['Quantity'] * data['UnitPrice']
data
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
customer_data = data.groupby('CustomerID').agg({
    'Quantity': 'sum',
    'InvoiceNo': 'count',
    'TotalAmount': 'sum'
}).reset_index()

customer_data.columns = ['CustomerID', 'TotalQuantity', 'PurchaseFrequency', 'TotalSpent']
customer_data.head()
```

```python
X = customer_data[['TotalQuantity', 'PurchaseFrequency', 'TotalSpent']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data['Class_Label'] = kmeans.fit_predict(X_scaled)

customer_data.head()
```

```python
customer_data.groupby('Class_Label').mean()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(customer_data['TotalSpent'], customer_data['PurchaseFrequency'],
            c=customer_data['Class_Label'], cmap='viridis')
plt.xlabel("Total Spent")
plt.ylabel("Purchase Frequency")
plt.title("Customer Segmentation using K-Means")
plt.colorbar(label='Class Label')
plt.show()
```

```python
snapshot_date = data['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm_data = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',                                    # Frequency
    'TotalAmount': 'sum'                                     # Monetary
}).reset_index()

rfm_data.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm_data.head()
```

```python
final_data = customer_data.merge(rfm_data, on='CustomerID')
final_data.head()
```

```python
X_generalized = final_data[
    ['TotalQuantity', 'PurchaseFrequency', 'TotalSpent',
     'Recency', 'Frequency', 'Monetary']
]
```

```python
scaler = StandardScaler()
X_generalized_scaled = scaler.fit_transform(X_generalized)
```

```python
kmeans_general = KMeans(n_clusters=3, random_state=42)
final_data['Generalized_Class_Label'] = kmeans_general.fit_predict(X_generalized_scaled)

final_data.head()
```

```python
final_data.groupby('Generalized_Class_Label').agg({
    'TotalQuantity': 'mean',
    'PurchaseFrequency': 'mean',
    'TotalSpent': 'mean',
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
})
```

```python
cluster_names = {
    0: "Low Value / Inactive",
    1: "Potential Customers",
    2: "Loyal High-Value",
    3: "Regular Customers"
}

final_data['Cluster_Name'] = final_data['Generalized_Class_Label'].map(cluster_names)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))

plt.bar(final_data['Cluster_Name'],
        final_data['PurchaseFrequency'])

plt.xlabel("Customer Segments")
plt.ylabel("Average Buying Frequency")
plt.title("Customer Segmentation Based on Buying Behavior")
plt.xticks(rotation=20)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_generalized_scaled,
                             final_data['Generalized_Class_Label'])

sil_score
```

```python
print(f"Silhouette Score for Generalized K-Means Clustering: {sil_score:.3f}")
```
